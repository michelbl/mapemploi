## Load active population at commune level

## Load coordinates of communes

```python
csv_communes = pd.read_csv('villes_france.csv', sep=',', header=None, dtype=str)
```

```python
pos_communes = csv_communes[[10, 19, 20]].copy()
```

```python
pos_communes.columns = ['code_insee', 'longitude', 'latitude']
```

```python
pos_communes
```

```python
pos_communes.latitude = pos_communes.latitude.map(float)
pos_communes.longitude = pos_communes.longitude.map(float)
```

```python
pos_communes = pos_communes[pos_communes.latitude != 0]
pos_communes = pos_communes[pos_communes.longitude != 0]
```

```python
def geo2carto(row):
    latitude = row['latitude']
    longitude = row['longitude']
    
    x, y = geo.geo2carto(latitude, longitude)
    
    return x, y
```

```python
x_y_couples = pos_communes.apply(geo2carto, axis=1)

pos_communes['X'] = x_y_couples.map(lambda xy: xy[0])
pos_communes['Y'] = x_y_couples.map(lambda xy: xy[1])
```

```python
pos_communes.drop(['longitude', 'latitude'], axis=1, inplace=True)
```

```python
pos_communes.to_csv('coordinates_communes.csv', index=False)
```

```python
pos_communes['active_population'] = 0.
```

```python
code_insee_groups = pos_communes.groupby('code_insee').groups
```

```python
i = 0
for commune, active_pop in active_dict_commune.items():
    if commune not in code_insee_groups:
        continue
    match_indexes = code_insee_groups[commune]
    assert len(match_indexes) <= 1
    if len(match_indexes) == 1:
        match_index = match_indexes[0]
        pos_communes.ix[match_index, 'active_population'] = active_pop
    i += 1
    if i%1000 == 0:
        print(i)
```

```python
pos_communes.active_population = pos_communes.active_population.map(float)
```

```python
pos_communes = pos_communes[pos_communes.active_population > 0]
```

```python
len(pos_communes)
```

```python
commune_data = {
    row['code_insee']: {
        'X': row['X'],
        'Y': row['Y'],
        'active_population': row['active_population'],
    }
    for _, row in pos_communes.iterrows()
}
```
