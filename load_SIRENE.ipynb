```python
import pandas as pd
import numpy as np
import os
import pickle
pd.set_option('display.max_columns', 150)

import geo
```

```python
TEFET2num_dict = {
    'NN': 0,
    '00': 0,
    '01': 1.5,
    '02': 4,
    '03': 7.5,
    '11': 15,
    '12': 35,
    '21': 75,
    '22': 150,
    '31': 225,
    '32': 375,
    '41': 750,
    '42': 1500,
    '51': 3500,
    '52': 7500,
    '53': 15000,
}

def TEFET2num(TEFET):
    return TEFET2num_dict[TEFET]
```

```python
EFETCENT2num_dict = {
    'NN': 0,
    '00': 0,
    '0': 0,
    '1': 1.5,
    '3': 4,
    '6': 7.5,
    '10': 15,
    '20': 35,
    '50': 75,
}

def EFETCENT2num(EFETCENT):
    if EFETCENT in EFETCENT2num_dict:
        return EFETCENT2num_dict[EFETCENT]
    
    EFETCENT_int = int(EFETCENT)
    assert not EFETCENT_int%100
    return EFETCENT_int + 50
```

```python
def merge_TEFET_EFETCENT(row):
    if row['EFETCENT_int'] == 0:
        return row['TEFET_int']
    
    return row['EFETCENT_int']
```

```python
csv_data_list = []
i = 0
filename_list = os.listdir("sirene")
filename_list.sort()
for filename in filename_list:
    print(filename)
    filename_full = 'sirene/' + filename
    
    csv_data = pd.read_csv(filename_full, dtype=str)
    
    if len(csv_data) == 0:  # departement 96
        continue
        
    csv_data = csv_data[['longitude', 'latitude', 'TEFET', 'EFETCENT']].copy()
    
    csv_data['TEFET_int'] = csv_data.TEFET.map(TEFET2num)
    csv_data['EFETCENT_int'] = csv_data.EFETCENT.map(EFETCENT2num)
    csv_data['effectif'] = csv_data.apply(merge_TEFET_EFETCENT, axis=1)
    csv_data.drop(['TEFET', 'EFETCENT', 'TEFET_int', 'EFETCENT_int'], axis=1, inplace=True)
    
    csv_data.to_csv('sirene_light_{}.csv'.format(i), index=False)
    i += 1
    
assert(i == 99)
```

## Geo to carto

```python
def geo2carto(row):
    latitude = float(row['latitude'])
    longitude = float(row['longitude'])
    
    x, y = geo.geo2carto(latitude, longitude)
    
    return x, y
```

```python
for i in range(99):
    print(i)
        
    csv_data = pd.read_csv('tmp_files/sirene_light_{}.csv'.format(i))
        
    x_y_couples = csv_data.apply(geo2carto, axis=1)
    
    csv_data['X'] = x_y_couples.map(lambda xy: xy[0])
    csv_data['Y'] = x_y_couples.map(lambda xy: xy[1])
    
    csv_data[['X', 'Y', 'effectif']].to_csv('tmp_files/sirene_xy_{}.csv'.format(i), index=False)
```

## Merge

```python
data_sirene = None
for i in range(99):
    print(i)
        
    csv_data = pd.read_csv('tmp_files/sirene_xy_{}.csv'.format(i))
    if data_sirene is None:
        data_sirene = csv_data
    else:
        data_sirene = pd.concat([data_sirene, csv_data])
```
